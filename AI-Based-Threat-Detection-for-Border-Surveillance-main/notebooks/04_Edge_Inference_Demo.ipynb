```python
import cv2, time, os
from ultralytics import YOLO

SOURCE = 'data/raw/sample.mp4'
CONF = 0.35
WEIGHTS_ONNX = 'weights/border_yolov8n.onnx'
WEIGHTS_PT = 'runs/train/exp/weights/best.pt'

def run_infer(source, conf=0.35):
    weights = WEIGHTS_ONNX if os.path.exists(WEIGHTS_ONNX) else WEIGHTS_PT
    print('Loading:', weights)
    model = YOLO(weights)
    cap = cv2.VideoCapture(source); assert cap.isOpened(), f'Cannot open {source}'
    prev = time.time()
    while True:
        ok, frame = cap.read()
        if not ok: break
        res = model.predict(source=frame, conf=conf, verbose=False)[0]
        annotated = res.plot()
        now = time.time(); fps = 1.0/(now - prev + 1e-6); prev = now
        cv2.putText(annotated, f'FPS:{fps:.1f}', (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
    cap.release(); print('Done.')

run_infer(SOURCE, CONF)
```
