```python
import os
from ultralytics import YOLO

WEIGHTS_PT = 'runs/train/exp/weights/best.pt'
OUT_ONNX = 'weights/border_yolov8n.onnx'
OUT_ENGINE = 'weights/border_yolov8n.engine'

os.makedirs('weights', exist_ok=True)
model = YOLO(WEIGHTS_PT)

model.export(format='onnx', imgsz=640, half=True, dynamic=True, optimize=True, simplify=True)
if os.path.exists('weights/best.onnx'):
    os.replace('weights/best.onnx', OUT_ONNX)
print('Exported ONNX to', OUT_ONNX)

try:
    model.export(format='engine', imgsz=640, half=True, dynamic=False, workspace=2)
    if os.path.exists('weights/best.engine'):
        os.replace('weights/best.engine', OUT_ENGINE)
    print('Exported TensorRT engine to', OUT_ENGINE)
except Exception as e:
    print('TensorRT export skipped or failed:', e)
```
