```python
from pathlib import Path
import shutil, json

# adjust these if your layout differs
cfg_path = Path(r"D:\Shihab's files\AI-Based_Threat_Detection_for_Border_Surveillance\configs\borderline.json")
cfg = json.loads(Path(cfg_path).read_text(encoding="utf-8"))

src_onnx = (cfg_path.parent / cfg["weights_onnx"]).resolve()  # current ONNX
safe_onnx = (cfg_path.parent / "weights" / "_tmp_border.onnx").resolve()
safe_onnx.parent.mkdir(parents=True, exist_ok=True)
shutil.copy2(src_onnx, safe_onnx)

cfg["weights_onnx"] = str(safe_onnx)  # use the safe path with no apostrophe in file name
Path(cfg_path).write_text(json.dumps(cfg, indent=2), encoding="utf-8")
print("Updated config to use:", cfg["weights_onnx"])
```

```python
from pathlib import Path
import sys, subprocess, json

CFG_PATH = Path(r"D:\Shihab's files\AI-Based_Threat_Detection_for_Border_Surveillance\configs\borderline.json")  # absolute path

print("Notebook CWD:", Path(".").resolve())
print("Using config :", CFG_PATH)
cfg = json.loads(CFG_PATH.read_text(encoding="utf-8"))

def resolve(p):
    p = Path(p)
    return p if p.is_absolute() else (CFG_PATH.parent / p)

for k in ["weights_onnx","weights_pt","input_source","output_video"]:
    p = resolve(cfg[k])
    print(f"{k:12s} -> {p} | exists: {p.exists()}")

cmd = [sys.executable, str((Path("../src/borderline_infer.py")).resolve()), "--config", str(CFG_PATH)]
print("\nRunning:", " ".join(cmd))
proc = subprocess.run(cmd, capture_output=True, text=True)
print("\n--- STDOUT ---\n", proc.stdout)
print("\n--- STDERR ---\n", proc.stderr)
print("\nReturn code:", proc.returncode)
```
