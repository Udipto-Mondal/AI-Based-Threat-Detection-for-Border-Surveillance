```python
import os
from pathlib import Path
from ultralytics.utils import SETTINGS

# 1) Point Ultralytics to your actual datasets folder (use DOUBLE quotes for the ' in Shihab's)
DATASETS_DIR = Path("D:/Shihab's files/AI-Based_Threat_Detection_for_Border_Surveillance/datasets")

assert DATASETS_DIR.is_dir(), f"Missing folder: {DATASETS_DIR}"
SETTINGS.update({'datasets_dir': str(DATASETS_DIR)})
print("Ultralytics datasets_dir =", SETTINGS['datasets_dir'])

# 2) Sanity-check the expected subfolders
for p in [
    DATASETS_DIR / "VisDrone",
    DATASETS_DIR / "VisDrone/VisDrone2019-DET-train",
    DATASETS_DIR / "VisDrone/VisDrone2019-DET-val"
]:
    print(p, "OK" if p.exists() else "NOT FOUND")
```

```python
from ultralytics import YOLO

DATA_CFG  = 'VisDrone.yaml'       # built-in Ultralytics config (will convert VisDrone to YOLO)
BASE_MODEL = '../yolov8n.pt'      # because your notebook is in /notebooks and the .pt is at project root
EPOCHS    = 50
IMGSZ     = 640
BATCH     = 16

model = YOLO(BASE_MODEL)
results = model.train(data=DATA_CFG, epochs=EPOCHS, imgsz=IMGSZ, batch=BATCH,
                      project='runs/train', name='exp')
results
```
